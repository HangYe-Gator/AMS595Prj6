## 2. Implementing Gradient Descent Algorithm (60 pts)

Given a function $f(x) = \sqrt{x^2+5}$, $x\in \mathbf{R}$.

Fact: This function is L-smooth.
0. Find the gradient of this function and the minimum value of this function analytically.


1. Plot this function over the interval $[-5, 5]$.
2. Perform the Gradient Descent algorithm to find the minimum value of $f$ for $50$ iterations ($T$) with a step size of $1$ ($\alpha$). Use the last digit of your SB ID divided by 5 and then plus 1 as the initial guess (for example, if your ID ends with 9, your initial guess $x_0$ will be $9\div5 + 1= 2.8$).
3. Record the values of $x_k$ at the $k$-th iteration during GD and report $x_T$.
4. Plot the value of $f(x_k)$ v.s. the iteration number $k$.
5. For each of the step sizes 5, 3, 1, and 0.5, perform gradient descent and record the values of $x_k$ in each step $k$. Plot $f(x_{k-1}) - f(x_{k})$ v.s. $k$ for each step size.
```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x)
def f(x):
    return np.sqrt(x**2 + 5)

# Define the derivative of the function f(x), for gradient descent
def f_prime(x):
    return x / np.sqrt(x**2 + 5)

# Write a function to run the Gradient Descent Algorithm
def gradient_descent(x_0, step_size, num_iterations):
    x_k = [x_0]  # Initialize with the starting point
    for _ in range(num_iterations):
        x_k.append(x_k[-1] - step_size * f_prime(x_k[-1]))
    return np.array(x_k)

# Set up/assign initial values
x_0 = 9/5+1  # 115669099
num_iterations = 50  # Number of iterations
step_sizes = [5, 3, 1, 0.5]  # Different step sizes to test

# Run gradient descent for the step size alpha=1 as an example
alpha = 1
x_k_alpha1 = gradient_descent(x_0, alpha, num_iterations)
x_T_alpha1 = x_k_alpha1[-1]
print(f"x_T for step size {alpha}: {x_T_alpha1}")

# Plot the function f(x) over the interval [-5, 5]
x_vals = np.linspace(-5, 5, 500)
plt.plot(x_vals, f(x_vals), label='f(x)')
plt.title("Function Plot")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()
plt.legend()
plt.show()

```

```python
# Plot f(x_k) vs. iteration number for step size alpha=1
f_x_k = f(x_k_alpha1)
plt.plot(range(num_iterations + 1), f_x_k, marker='o', label=f'Alpha={alpha}')
plt.title("f(x_k) vs Iteration")
plt.xlabel("Iteration")
plt.ylabel("f(x_k)")
plt.grid()
plt.legend()
plt.show()
```

```python
# Initialize a list to store the results from different step sizes
results = {}

# Loop through the step sizes and perform GD and record results
for step_size in step_sizes:
    x_k = gradient_descent(x_0, step_size, num_iterations)
    f_x_k = f(x_k)
    results[step_size] = f_x_k
    print(f"x_T for step size {step_size}: {x_k[-1]}")

# Plot multiple curves in a single graph
plt.figure(figsize=(10, 6))
for step_size, f_x_k in results.items():
    plt.plot(range(num_iterations + 1), f_x_k, label=f'Alpha={step_size}')
plt.title("f(x_k) vs Iteration for Different Step Sizes")
plt.xlabel("Iteration")
plt.ylabel("f(x_k)")
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate synthetic data
np.random.seed(99)  # My sbu id is 115669099
X = 2 * np.random.rand(100, 2)  # Generate 100 random 2D data points
y = 4 + 4 * X[:, 0] + 0.5 * X[:, 1] + np.random.randn(100)  # Generate y values with noise

# Add a column of ones to X for the bias term (intercept) in the normal equation
X_b = np.c_[np.ones((100, 1)), X]  # Add a column of ones to X

# Find theta using the normal equation
theta = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

# Print the intercept and coefficients of the regression plane
intercept = theta[0]
coefficients = theta[1:]
print(f"Intercept: {intercept}")
print(f"Coefficients: {coefficients}")

# Plot the generated data and the regression plane
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot of the data points
ax.scatter(X[:, 0], X[:, 1], y, color='blue', label='Data Points')

# Generate a meshgrid for the regression plane
x0 = np.linspace(0, 2, 50)
x1 = np.linspace(0, 2, 50)
x0, x1 = np.meshgrid(x0, x1)
y_pred = intercept + coefficients[0] * x0 + coefficients[1] * x1

# Plot the regression plane
ax.plot_surface(x0, x1, y_pred, alpha=0.5, color='orange', label='Regression Plane')

# Label the axes
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('y')
plt.title('2D Linear Regression with Regression Plane')
plt.legend()
plt.show()
```

## 4. Logistic Regression with Numpy and Graphing with Matplotlib (20 pts)

In class, we implemented a logistic regression model to perform binary classification. Here, I provided the exact same code that we discussed in class, but the values for learning_rate and num_epochs are modified. I also removed all the comments. You task is to
1. Tune the Learning Rate and Number of Epochs: Your first task is to fine-tune the values for the learning_rate and num_epochs parameters. Your goal is to identify suitable values that will enable us to converge to a set of parameter values $\theta$ closely approximating the optimal parameter values $\theta^*$.
2. Code Explanation: To enhance code comprehension, please augment the code with meaningful comments. These comments should elucidate the purpose and functionality of each code segment, making it easier for readers to understand the logistic regression implementation.

By accomplishing these tasks, we aim to achieve a better understanding of the logistic regression model's behavior and its parameter optimization process.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sigmoid function, which maps real numbers to a (0, 1) range
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Generate synthetic data for binary classification
np.random.seed(595)  # Ensure reproducibility
X = np.random.rand(2, 100)  # Generate random 2D feature points
w_true = np.array([1.5, -2.5])  # True weights
b_true = 1.0  # True bias
probabilities = sigmoid(np.dot(w_true.T, X) + b_true)  # Generate probabilities using the sigmoid function
Y = (probabilities > 0.5).astype(int)  # Assign binary labels based on probabilities
X = X + 0.3 * np.random.rand(2, 100) - 0.1 * np.random.rand(2, 100)  # Add some noise to the data

# Split data into training and testing sets
X_train, X_test = X[:, :80], X[:, 80:]  # 80% for training, 20% for testing
Y_train, Y_test = Y[:80], Y[80:]  # Corresponding labels

# Initialize weights and bias for logistic regression
w = np.zeros(X.shape[0])  # Start with zero weights
b = 0.0  # Start with zero bias

# Set learning rate and number of epochs for gradient descent
learning_rate = 0.1  
num_epochs = 1000  

# Perform gradient descent to optimize weights and bias
for epoch in range(num_epochs):
    # Compute predictions for the training set
    A_train = sigmoid(np.dot(w.T, X_train) + b)
    
    # Compute gradients of the loss function w.r.t. weights and bias
    dJdw = np.dot(X_train, (A_train - Y_train).T) / len(Y_train)
    dJdb = np.mean(A_train - Y_train)
    
    # Update weights and bias using the gradients
    w -= learning_rate * dJdw
    b -= learning_rate * dJdb

# Evaluate the model on the training set
A_train = sigmoid(np.dot(w.T, X_train) + b)
predictions_train = (A_train > 0.5).astype(int)

# Evaluate the model on the testing set
A_test = sigmoid(np.dot(w.T, X_test) + b)
predictions_test = (A_test > 0.5).astype(int)

# Compute accuracy for both training and testing sets
train_accuracy = np.mean(predictions_train == Y_train)
test_accuracy = np.mean(predictions_test == Y_test)

# Print the results
print(f"Training Set Accuracy: {train_accuracy:.2f}")
print(f"Test Set Accuracy: {test_accuracy:.2f}")
```

```python
plt.scatter(X[0, :80], X[1, :80], c=Y[:80], cmap=plt.cm.Paired)
plt.xlim(-0.2, 1.2)
plt.ylim(-0.2, 1.2)
ax = plt.gca()

# Generate a meshgrid for plotting the decision boundary
xlim = ax.get_xlim()
ylim = ax.get_ylim()
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 50), np.linspace(ylim[0], ylim[1], 50))
Z = np.dot(w.T, np.c_[xx.ravel(), yy.ravel()].T) + b
Z = sigmoid(Z)
Z = Z.reshape(xx.shape)

# Plot the decision boundary
plt.contour(xx, yy, Z, colors='k', levels=[0.5], alpha=0.5, linestyles=['--'])
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Logistic Regression Decision Boundary (Training Data)')
plt.show()
```

```python

# Plot decision boundary on the testing data
plt.scatter(X[0, 80:], X[1, 80:], c=Y[80:], cmap=plt.cm.Paired)
plt.xlim(-0.2, 1.2)
plt.ylim(-0.2, 1.2)
ax = plt.gca()

# Reuse the meshgrid and decision boundary for the test data
Z = np.dot(w.T, np.c_[xx.ravel(), yy.ravel()].T) + b
Z = sigmoid(Z)
Z = Z.reshape(xx.shape)

# Plot the decision boundary
plt.contour(xx, yy, Z, colors='k', levels=[0.5], alpha=0.5, linestyles=['--'])
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Logistic Regression Decision Boundary (Testing Data)')
plt.show()
```
